```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error
```

```python
# 1. Load datasets
train = pd.read_csv(r"C:\Users\HP\Downloads\train.csv")
test = pd.read_csv(r"C:\Users\HP\Downloads\test.csv")
stores = pd.read_csv(r"C:\Users\HP\Downloads\stores.csv")
features = pd.read_csv(r"C:\Users\HP\Downloads\features.csv")
```

```python
# 2. Convert Date columns to datetime
train['Date'] = pd.to_datetime(train['Date'])
test['Date'] = pd.to_datetime(test['Date'])
features['Date'] = pd.to_datetime(features['Date'])
```

```python
# 3. Merge train with stores and features
train_merged = pd.merge(train, stores, on="Store", how="left")
train_merged = pd.merge(train_merged, features, on=["Store", "Date"], how="left")
```

```python
# 4. Merge test with stores and features
test_merged = pd.merge(test, stores, on="Store", how="left")
test_merged = pd.merge(test_merged, features, on=["Store", "Date"], how="left")
```

```python
# 5. Aggregate sales by week (sum across all stores/departments)
weekly_sales = train_merged.groupby(pd.Grouper(key="Date", freq="W"))['Weekly_Sales'].sum()
```

```python
# 6. Plot sales trend
plt.figure(figsize=(12,6))
weekly_sales.plot(title="Weekly Sales Trend")
plt.show()
```

```python
# 7. Check seasonality with rolling mean
rolling_mean = weekly_sales.rolling(window=12).mean()
plt.figure(figsize=(12,6))
weekly_sales.plot(label="Weekly Sales")
rolling_mean.plot(label="Rolling Mean (12 weeks)")
plt.legend()
plt.show()
```

```python
# 8. Split into train/test by time
train_data = weekly_sales[weekly_sales.index < '2012-01-01']
test_data = weekly_sales[weekly_sales.index >= '2012-01-01']
```

```python
# 9. Fit forecasting model (Exponential Smoothing)
model = ExponentialSmoothing(train_data, trend="add", seasonal=None)
fit = model.fit()
forecast_es = fit.forecast(len(test_data))
```

```python
# 10. Plot forecast vs actual
plt.figure(figsize=(12,6))
train_data.plot(label="Train")
test_data.plot(label="Test")
forecast_es.plot(label="Forecast", color="red")
plt.legend()
plt.show()
```

```python
# 11. Compute error metrics
mae = mean_absolute_error(test_data, forecast_es)
mape = mean_absolute_percentage_error(test_data, forecast_es)
print("MAE:", mae)
print("MAPE:", mape)
```

```python
# 12. Export forecast results
forecast_es.to_csv("forecast_results.csv", header=["Forecast"])
```
